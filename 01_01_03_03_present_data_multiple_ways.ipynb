```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd  
%matplotlib inline

titanic = sns.load_dataset('titanic')

titanic.head()
```

Choose one variable and plot that variable four different ways.

```python
# what do I do about NaN???

sns.distplot(titanic['age'])
```

```python
sns.kdeplot(titanic['age'])
```

```python
sns.rugplot(titanic['age'])
```

```python
sns.distplot(titanic['fare'])
```

Choose two continuous variables, and plot them three different ways.

```python
# replot() ????

sns.relplot(x="age", y="fare", data=titanic);
```

```python
sns.jointplot(x='age',y='fare',data=titanic,kind='scatter')
```

```python
sns.jointplot(x='age',y='fare',data=titanic,kind='reg')
```

Choose one continuous variable and one categorical variable, and plot them six different ways.

```python
sns.barplot(x='class',y='fare',data=titanic)
```

```python
sns.factorplot(x='class',y='fare',data=titanic,kind='bar')
```

```python
# catplot()( ?????

sns.catplot(x="class", y="fare", kind="box", data=titanic);
```

```python
sns.boxplot(x="class", y="fare", data=titanic,palette='rainbow')
```

```python
sns.violinplot(x="class", y="fare", data=titanic,palette='rainbow')
```

```python
sns.boxplot(x="class", y="fare", hue="sex",data=titanic, palette="rainbow")
```

```python
sns.violinplot(x="class", y="fare", data=titanic,hue='sex',palette='rainbow')
```

```python
sns.stripplot(x="class", y="fare", data=titanic,jitter=True,hue='sex',palette='rainbow',split=True)
```
